### Bryan
- **Access to electricity (% of population):** En español, **Acceso a la electricidad (% de población)**, mide el porcentaje de la población que tiene acceso a la electricidad. Este dato es crucial para evaluar el desarrollo y la calidad de vida en una región, ya que el acceso a la electricidad es esencial para actividades cotidianas, educación, salud y economía.

Puede observarse que los países con un mayor porcentaje de acceso a la electricidad son:
1. Costa Rica
2. El Salvador
3. Belize

Mientras que los paises con menos acceso a electricidad son:
1.  Nicaragua
2. Honduras

(Datos tomando como periodo de tiempo [2010 - 2022])

```python
# seleccione las filas que contienen el indicador "Access to electricity (% of population" 
dfAccessElectricity = df[df['Indicador'].str.contains('Access to electricity (% of population)', na=False, regex=False)] 
# eliminar la columna indicador
dfAccessElectricity  = dfAccessElectricity.drop(columns=['Indicador'])
# elegimos la columna " País" como index y eliminamos la columna "País"
dfAccessElectricity .set_index('País', inplace=True)
# mostramos los datos 
dfAccessElectricity 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrar por un indicador específico
dfAccessElectricity = df[df['Indicador'] == "Access to electricity (% of population)"]

# Reemplazar valores no numéricos (como '..') por NaN
dfAccessElectricity.replace('..', pd.NA, inplace=True)

# Convertir columnas de años a numérico
dfAccessElectricity.iloc[:, 2:] = dfAccessElectricity.iloc[:, 2:].apply(pd.to_numeric, errors='coerce')

# Eliminar la columna "Indicador"
dfAccessElectricity = dfAccessElectricity.drop(columns=['Indicador'])

# Establecer la columna "País" como índice
dfAccessElectricity.set_index('País', inplace=True)

# Crear el gráfico
plt.figure(figsize=(12, 6))

# Iterar sobre cada país para graficar su línea
for pais in dfAccessElectricity.index:
    plt.plot(
        dfAccessElectricity.columns,          # Años (columnas)
        dfAccessElectricity.loc[pais],       # Valores del país (filas)
        marker='o', 
        label=pais
    )

# Configuración del gráfico
plt.title('Acceso a electricidad por país (% de población) [2010 - 2022]', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Acceso a electricidad (%)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='País', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()  # Ajustar el espacio para que todo quepa bien
plt.show()
```

Este gráfico ilustra la evolución del porcentaje de acceso a la electricidad por país en Centroamérica entre 2010 y 2022.

En términos generales, todos los países muestran una tendencia creciente en el acceso a la electricidad, lo que refleja un avance significativo en la cobertura de este servicio básico. Costa Rica destaca de manera sobresaliente al mantener una cobertura cercana al 100%, alcanzando el 100% desde 2021 y manteniéndolo durante 2022.

El Salvador también logró llegar al 100% en 2022, uniéndose a Costa Rica en términos de cobertura total. En cambio, Honduras, que se encuentra entre los países con menos acceso, experimentó una caída notable en 2017, cuando su porcentaje de acceso pasó del 91.6% al 86.5%. Esta disminución podría estar relacionada con diversos factores, como problemas de infraestructura, interrupciones en el suministro eléctrico o desafíos en las políticas energéticas nacionales de ese año.

Nicaragua, por otro lado, sigue siendo el país con el menor porcentaje de acceso a la electricidad, aunque muestra una mejora constante a lo largo del tiempo. A pesar de esto, su cobertura sigue siendo inferior al 90%, lo que subraya la necesidad de continuar trabajando en el fortalecimiento de su infraestructura eléctrica para garantizar un acceso más equitativo.

En resumen, aunque hay avances en todos los países, las disparidades en el acceso a la electricidad continúan siendo notorias, con Costa Rica y El Salvador a la vanguardia, y Nicaragua enfrentando los mayores desafíos en este ámbito.

```python
import matplotlib.pyplot as plt
# Reemplazar '..' por NaN
dfAccessElectricity.replace('..', pd.NA, inplace=True)

# Convertir todas las columnas (excepto el índice) a valores numéricos
dfAccessElectricity = dfAccessElectricity.apply(pd.to_numeric, errors='coerce')

# Calcular la media de cada fila (promedio por país)
dfAccessElectricity['Media'] = dfAccessElectricity.mean(axis=1)

# Crear el gráfico de barras horizontales para la media de cada país
dfAccessElectricity['Media'].plot(kind='barh', figsize=(10, 6))

# Añadir etiquetas y título
plt.xlabel('Promedio de acceso a electricidad (%)')
plt.ylabel('País')
plt.title('Promedio de acceso a electricidad por país (% de población) [2010 - 2022]')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Calcular describe() considerando los países
describe_por_pais = dfAccessElectricity.T.describe().T

# Mostrar el resultado
print(describe_por_pais)

```

- **Net migration:** En español, **migración neta**, es el número de inmigrantes menos el número de emigrantes, incluidos ciudadanos y no ciudadanos, durante un período determinado. Si el resultado es negativo, significa que hay una disminución en la cantidad de inmigrantes y un aumento en la cantidad de emigrantes. Si el resultado es positivo, significa que hay un aumento en la cantidad de inmigrantes y una disminución en la cantidad de emigrantes.
    - Emigrante es aquel que abandona el país del cual es originario para establecerse en uno nuevo en el extranjero. 
    - Inmigrante es aquel que llega a un país distinto de aquel del que es originario con el objetivo de radicarse en él.

Es de tener en cuenta que la migración neta se calcula como la cantidad de inmigrantes menos la cantidad de emigrantes, lo que significa que si el resultado es negativo, hay más personas saliendo de su país de origen que personas entrando a dicho país, mientras que si el resultado es positivo, hay más personas entrando a ese país que personas saliendo a dicho país.

Esto puede indicar que hay o puede haber una sobrepoblación en este país, ya que el área territorial de El Salvador es de 21,041 km² y su densidad poblacional es alta en comparación con la densidad de población del resto de países de la región centroamericana, quienes tienen más área territorial. Mientras que en otros países, como Belice, Nicaragua y Panamá, la densidad poblacional es mucho menor, lo que sugiere que hay una baja población en estos países. En general, es importante tener en cuenta la densidad poblacional para entender la evolución de la población en un país determinado.
